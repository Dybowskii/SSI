```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
def happy_face():
    x_eyes = [-1,0,1]
    y_eyes = [1,0,1]
    x_smile = np.linspace(-1, 1, 100)
    y_smile = np.sin(x_smile*2-1.6)
    plt.scatter(x_eyes,y_eyes)
    plt.plot(x_smile,y_smile, color='yellow')
    angle = np.linspace( 0 , 2 * np.pi , 150 ) 
    radius = 2
    x = radius * np.cos( angle ) 
    y = radius * np.sin( angle ) 
    plt.plot(x, y, color='red')
    plt.legend(['funkcja okręgu','sinus','punkty'])

happy_face()
plt.show()
```

class(1=Setosa,2=Versicolour,3=Virginica)

```python
df_class_setosa = probki.loc[probki['class']==1]
df_class_versicolour = probki.loc[probki['class']==2]
df_class_virginica = probki.loc[probki['class']==3]
```

```python
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 1)
plt.scatter(df_class_setosa['petal_length_in_cm'], df_class_setosa['petal_width_in_cm'], label='Setosa', marker='x')
plt.scatter(df_class_versicolour['petal_length_in_cm'], df_class_versicolour['petal_width_in_cm'], label='Versicolour', marker='p')
plt.scatter(df_class_virginica['petal_length_in_cm'], df_class_virginica['petal_width_in_cm'], label='Virginica', marker = 's')
plt.xlabel('petal_length_in_cm')
plt.ylabel('petal_width_in_cm')
plt.legend()
plt.subplot(2, 2, 2)
plt.scatter(df_class_setosa['sepal_width_in_cm'], df_class_setosa['petal_width_in_cm'], label='Setosa', marker='x')
plt.scatter(df_class_versicolour['sepal_width_in_cm'], df_class_versicolour['petal_width_in_cm'], label='Versicolour', marker='p')
plt.scatter(df_class_virginica['sepal_width_in_cm'], df_class_virginica['petal_width_in_cm'], label='Virginica', marker = 's')
plt.xlabel('sepal_width_in_cm')
plt.ylabel('petal_width_in_cm')
plt.legend()
plt.subplot(2, 2, 3)
plt.scatter(df_class_setosa['sepal_length_in_cm'], df_class_setosa['petal_width_in_cm'], label='Setosa', marker='x')
plt.scatter(df_class_versicolour['sepal_length_in_cm'], df_class_versicolour['petal_width_in_cm'], label='Versicolour', marker='p')
plt.scatter(df_class_virginica['sepal_length_in_cm'], df_class_virginica['petal_width_in_cm'], label='Virginica', marker = 's')
plt.xlabel('sepal_length_in_cm')
plt.ylabel('petal_width_in_cm')
plt.legend()
plt.subplot(2, 2, 4)
plt.scatter(df_class_setosa['sepal_width_in_cm'], df_class_setosa['petal_length_in_cm'], label='Setosa', marker='x')
plt.scatter(df_class_versicolour['sepal_width_in_cm'], df_class_versicolour['petal_length_in_cm'], label='Versicolour', marker='p')
plt.scatter(df_class_virginica['sepal_width_in_cm'], df_class_virginica['petal_length_in_cm'], label='Virginica', marker = 's')
plt.xlabel('sepal_width_in_cm')
plt.ylabel('petal_length_in_cm')
plt.legend()
plt.show()
```
